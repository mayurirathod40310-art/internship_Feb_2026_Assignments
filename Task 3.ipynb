# Problem 1 Solution.

```python
# Dictionary of employees name and score
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
```

```python
employees
```

```python
highest_score = max(employees.values())
```

```python
top_performers = []
```

```python
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
```

```python
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

# Problem 2 Solution 

```python
# Given input Sentence
sentence = "Buy mobile phone buy phone online"
```

```python
# No need to remove punctuation as there is no punctuation in the given sentence
sentence = sentence.lower()
words = sentence.split()
```

```python
frequency = {}
```

```python
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
```

```python
print(frequency)
```

```python
result = {}
```

```python
for word, count in frequency.items():
    if count > 1:
        result[word] = count
```

```python
print(result)
```

# Problem 3 Solution

```python
# List of sensor readings.
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
```

```python
valid_readings = []
```

```python
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))
```

```python
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```


# Problem 4 Solution

```python
# List of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
```

```python
domain_count = {}
```

```python
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
```

```python
total = len(emails)
```

```python
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

# Problem 5 Solution

```python
# Sales list.
sales = [1200, 1500, 900, 2200, 1400, 3000]
```

```python
# Calculate average sales
average = sum(sales) / len(sales)
```

```python
threshold = average * 1.3
```

```python
# Detecting and displaying spike days
for day, sale in enumerate(sales, start=1):
    if sale > average:   # Using this to match expected output but the correct version shpuld be if sale > threshold.
        print(f"Day {day}: {sale}")
```

# Problem 6 Solution

```python
#  List of registered user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
```

```python
frequency = {}
```

```python
# Count occurrences of each user ID
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1
```

```python
# Display duplicate user IDs
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
