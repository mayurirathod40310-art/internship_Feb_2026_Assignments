# Problem 2 Solution

```python
# Creating dictionary of employees and their salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
```

```python
highest_employee = max(employees, key=employees.get)
```

# Problem 3 solution

```python
# List of numbers
values = [34, 78, 12, 56, 99, 23]
```

```python
maximum = values[0]
minimum = values[0]
```

```python
for num in values:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
```

```python
print("List:", values)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

# Problem 5 solution

```python
# List of attendence
attendance_record = ["P", "A", "P", "P", "A", "P"]
```

```python
total_days = len(attendance_record)
```

```python
present_days = 0
for status in attendance_record:
    if status == "P":
        present_days += 1
```

```python
attendance_percentage = (present_days / total_days) * 100
```

```python
print("Attendance Percentage:", attendance_percentage)
```

```python
# To round figure upto 2 decimal palaces 
print("Attendance Percentage:", round(attendance_percentage, 2))
```

# Problem 7 Solution

```python
text = "python"
```

```python
char_count = {}
```

```python
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
```

```python
print(char_count)
```
