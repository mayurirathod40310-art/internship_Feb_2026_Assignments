# Problem 1 Solution 

```python
#  Mobile Recharge Validation Function 
def recharge_validation(amount,Plan):
    if amount < 50:
        return "Invalid Amount"
    if Plan not in [199,299,399,599]:
        return "Invalid Plan"
    return "Recharge Successful !"
# User input 
while True:
    amount = int(input("Enter Amount:"))
    Plan = int(input("Enter Plan(199,299,399,599):"))
    result=recharge_validation(amount,Plan)
    print(result)

    if result=="Recharge Successful !":
        break
    else:
        print("Please try again.\n")
```


# Problem 2 Solution

```python
# Function 
def inventory_system(products):
    
    for item, stock in products.items():
        
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


# User input
products = {}

n = int(input("Enter number of products: "))

for i in range(n):
    name = input("Enter product name: ")
    quantity = int(input("Enter stock quantity: "))
    products[name] = quantity   # adding to dictionary

print("\nInventory Status:\n")

inventory_system(products)
```

# Problem 3 Solution

```python
# Function
def student_result(marks):
    
    total = 0
    
    for num in marks:
        total += num
    
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    if average >= 50:
        print("Pass")
    else:
        print("Fail")


# User input
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = int(input("Enter marks: "))
    marks.append(mark)

student_result(marks)
```

# Problem 4 Solution

```python
# Function to calculate fare .
def calculate_fare(distance, peak):
    
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (per_km_rate * distance)
    
    if peak == "yes":
        fare = fare * 1.25   # 25% extra
    
    return fare


while True:
    
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    total_fare = calculate_fare(distance, peak)
    
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    
    if retry != "yes":
        break
```

# Problem 5 Solution

```python
def attendance_eligibility(attendance):
    
    present_count = 0
    
    for status in attendance:
        if status == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance)) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = []

n = int(input("Enter total working days: "))

for i in range(n):
    status = input("Enter attendance (P/A): ").upper()
    attendance.append(status)

result = attendance_eligibility(attendance)

print(result)
```

# Problem 6 Solution

```python
# Password strength checking function.
def password_checker(password):
    
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8"
    
    has_digit = False
    has_special = False
    
    for char in password:
        
        if char.isdigit():
            has_digit = True
        
        if char in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
        

# User input
password = input("Enter password: ")

result = password_checker(password)

print(result)
```
