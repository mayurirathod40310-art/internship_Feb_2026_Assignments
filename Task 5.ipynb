# Problem 1: Social Media – Post Engagement Analyzer Solution

```python
def analyze_engagement(likes_list):
    # Initialize total likes to 0
    total_likes = 0
     # Loop through the list to calculate total likes
    for like in likes_list:
        total_likes += like
     # Check if post is viral or normal
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)
likes_data = [200, 300, 150, 400]   
analyze_engagement(likes_data)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker Solution

```python
def check_rainfall(rainfall_list, required_level):
    # Initialize total rainfall
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)
    
    # Check rainfall adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # results
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)
seasonal_rainfall = [70, 75, 68, 80, 67]
required_rainfall = 70
check_rainfall(seasonal_rainfall, required_rainfall)
```

# Problem 4: Social Media – Duplicate Account Detection Solution

```python
def detect_duplicate_accounts(usernames):
    # Converting  list to set to remove duplicates
    unique_usernames = set(usernames)
     # Compare lengths to detect duplicates
    if len(usernames) != len(unique_usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

user_list = ["noor123", "amit45", "noor123", "rahul78"]
detect_duplicate_accounts(user_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter Solution

```python
def filter_premium_crops(price_list):
    # Create an empty list to store premium crops
    premium_crops = []
     # Loop through price list
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    print("Premium Crops:", premium_crops)
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

# Problem 7: System Monitoring – Application Health Checker Solution

```python
def check_application_health(error_count):
    # Determine system status based on error count
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
     # Display system health status
    print("Error Count:", error_count)
    print("System Status:", status)
errors = 7
check_application_health(errors)
```

# Problem 9: E-Learning – Student Attendance Eligibility System Solution

```python
def check_exam_eligibility(attendance_list):
    # Initialize present days counter
    present_days = 0 
    # Count present days using loop
    for day in attendance_list:
        if day == 1:
            present_days += 1
      # Calculate attendance percentage
    attendance_percentage = (present_days / len(attendance_list)) * 100
    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)
attendance_data = [1, 1, 0, 1, 1]  
check_exam_eligibility(attendance_data)
```

# Problem 10: Smart Electricity Bill Analyzer Solution

```python
def calculate_electricity_bill(units):
    # Calculate bill based on slab system
    
    if units <= 100:
        bill = units * 3
    
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Classify usage level
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    # Return both bill and usage status
    return bill, usage_status
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", status)
```
