# Problem 1: Smart Light Controller Solution

```python
# Creating Class.
class Light:
    def __init__(self, name,status):
        self.name=name
        self.status=status

    def display_status(self):
        print(self.name, "is",self.status)
    
# Creating object
L1=Light(" Bedroom light","ON")
# SHOW RESULT
L1.display_status()  # Calling method
```

# Problem 2: Employee ID Card System Solution 


```python
# Creating class.

class Employee:
    def __init__(self,name,ID,department):
        self.name=name
        self.ID=ID
        self.department=department

    def display_details(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.ID)
        print("Department:",self.department)

# Creating object 
E1=Employee(" Rahul","EMP102", "IT")

# Calling method
E1.display_details()
```

# Problem 3: Mobile Contact Record Solution

```python
# Creating class

class contact :
    def __init__(self,name,phone_no):
        self.name=name
        self.phone_no=phone_no

    def display_contact_info(self):
        if len(self.phone_no)!=10:
            print("invalid Phone Number")
        else:
            print ("Contact Saved")
            print ("Name:",self.name)
            print ("Phone:",self.phone_no)

# Creating object
C1=contact("Anita","9876543210")
# Calling method
C1.display_contact_info()
```

# Problem 4: Product Price Tag Generator Solution

```python
# Creating class.
class product:
    def __init__(self,name, price):
        self.name=name
        self.price=price

    def display_price(self):
        print("Product:",self.name)
        print("Price:₹",self.price)
# Creating object 
P1=product("Headphones","2499")
# calling method
P1.display_price()
```

# Problem 5: Movie Rating Display System Solution

```python
# Creating class:
class movie:
    def __init__(self,name,rating):
         self.name=name
         self.rating=rating
    def display_details(self):
        print("Movie:",self.name)
        print("Rating",self.rating,"/ 5")
# Creating object 
M1=movie("Inception",4.8)

# Calling method
M1.display_details()
```

 # Problem 6: Delivery Address Manager Solution

```python
# Creating class
class address :
    def __init__(self,name,address):
        self.name=name
        self.address=address

    def display_delivery_details(self):
            print ("Delivery Details")
            print ("Name:",self.name)
            print ("Address:",self.address)

# Creating object
A1=address("Suman","Hydrabad")
# Calling method
A1.display_delivery_details()
```
