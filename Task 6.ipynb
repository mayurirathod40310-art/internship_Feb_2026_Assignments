# Problem 1- Smart Parking Lot Management System Solution

```python
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0
    
    for log in vehicle_logs:
        
        if log == "IN":
            current_parked += 1
            
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
        
        if current_parked > peak_usage:
            peak_usage = current_parked
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    
    if current_parked > capacity:
        print("Parking Status: Exceeded Capacity ⚠️")
    elif current_parked == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
```

```python
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

# Problem 2 - Online Food Delivery Time Estimator Solution

```python
def estimate_delivery_time(distance, traffic_level, weather):
    
    # Base time: 5 minutes per km
    base_time = distance * 5
    
    # Traffic delay
    if traffic_level == "Low":
        traffic_delay = 5
    elif traffic_level == "Medium":
        traffic_delay = 8
    elif traffic_level == "High":
        traffic_delay = 10
    else:
        traffic_delay = 0
    
    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 5
    elif weather == "Stormy":
        weather_delay = 15
    else:
        weather_delay = 0
    
    total_time = base_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", total_time, "minutes")
```

```python
distance = 8
traffic_level = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic_level, weather)
```

# Problem 3 - Movie Theatre Seat Occupancy Analyzer Solution

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    
    # Count total booked seats
    total_booked = len(booked_seats)
    
    # Calculate occupancy percentage
    occupancy_percentage = (total_booked / total_seats) * 100
    print("Occupancy:", round(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional show")
        
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
        
    elif occupancy_percentage >= 50:
        print("Show Status: Moderate")
        
    else:
        print("Show Status: Seats Available")
```

```python
 total_seats = 200
 booked_seats = [1] * 150   # 150 booked seats

 analyze_seat_occupancy(total_seats, booked_seats)
```

# Problem 4 - Cloud Server Load Classification System Solution

```python
def classify_server_load(cpu_readings):
    
    # Calculate total CPU usage using loop
    total = 0
    for reading in cpu_readings:
        total += reading
    
    # Calculate average CPU usage
    average_cpu = total / len(cpu_readings)
    
    # Display average (rounded)
    print("Average CPU Load:", round(average_cpu), "%")
    
    # Classify server status
    if average_cpu < 50:
        print("Server Status: Normal")
        
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")
        
    else:
        print("Server Status: Critical")
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

# Problem 5 - Smart Classroom Resource Usage Monitor Solution

```python
def monitor_resource_usage(resource_usage):
    
    overused_resources = []
    for resource, hours in resource_usage.items():
        
        # Check if usage exceeds 8 hours
        if hours > 8:
            overused_resources.append(resource)
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


#  Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function Call
monitor_resource_usage(resource_usage)
```

#  Problem 6 - Online Event Registration Capacity Controller Solution

```python
def manage_event_registration(capacity, registrations):
    
    confirmed = 0
    waitlisted = 0
    for i in range(registrations):
        
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    # Determine registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


#  Input
capacity = 100
registrations = 105

# Function Call
manage_event_registration(capacity, registrations)
```
